```python
from sklearn.cluster import KMeans
wcss = []
k_range = range(1, 15)
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['sepal_length', 'petal_length']])
    wcss.append(km.inertia_)
```

```python
plt.plot(k_range,wcss)
plt.title("elbow method")
plt.xlabel("k")
plt.ylabel("wcss")
```

```python
Kmeans=cluster.KMeans(n_clusters=5)
```
